# 📊 Alura Store - Análise de Dados com Python

Notebook criado para prática no **Google Colab** utilizando conceitos básicos de Python aplicados à análise de dados.

Conceitos aplicados:
- Variáveis
- Listas
- Loops (for)
- Condicionais (if / else)
- Funções

Objetivo: analisar dados simples de vendas da loja fictícia **Alura Store** e gerar alguns insights.

## 1️⃣ Criando os dados da loja

```python
produtos = ['Notebook', 'Mouse', 'Teclado', 'Monitor', 'Headset']
precos = [3500, 80, 150, 900, 200]
avaliacoes = [4.8, 4.5, 4.2, 4.6, 4.1]
quantidades = [5, 20, 15, 7, 10]

print('Dados carregados com sucesso!')
```

## 2️⃣ Calculando faturamento total

```python
faturamento_total = 0

for i in range(len(produtos)):
    faturamento = precos[i] * quantidades[i]
    faturamento_total += faturamento

print('Faturamento total da loja:', faturamento_total)
```

## 3️⃣ Verificando avaliação dos produtos

```python
for i in range(len(produtos)):
    if avaliacoes[i] >= 4.5:
        print(produtos[i], 'é bem avaliado ⭐')
    else:
        print(produtos[i], 'tem avaliação média')
```

## 4️⃣ Criando uma função para calcular faturamento

```python
def calcular_faturamento(preco, quantidade):
    return preco * quantidade
```

## 5️⃣ Usando a função na análise

```python
for i in range(len(produtos)):
    total = calcular_faturamento(precos[i], quantidades[i])
    print(produtos[i], 'faturou:', total)
```

## 6️⃣ Produto com maior faturamento

```python
maior_venda = 0
produto_top = ''

for i in range(len(produtos)):
    vendas = precos[i] * quantidades[i]
    if vendas > maior_venda:
        maior_venda = vendas
        produto_top = produtos[i]

print('Produto com maior faturamento:', produto_top)
```

## 📈 Conclusão

Com essa análise simples conseguimos:
- Calcular o faturamento total
- Identificar produtos mais bem avaliados
- Descobrir qual produto gera mais receita

Esses são conceitos fundamentais utilizados no dia a dia de um **Analista de Dados**.
